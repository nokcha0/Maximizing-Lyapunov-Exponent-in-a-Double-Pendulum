# I. Introduction

The purpose of this is to study the motion of a double pendulum. More particularly, the optimal starting angles that are needed for the highest entropy found using Lyapunov’s exponent. However, before this, we must first understand the motion of a double pendulum (Figure 1-3).

<img src=https://upload.wikimedia.org/wikipedia/commons/thumb/7/78/Double-Pendulum.svg/170px-Double-Pendulum.svg.png align="center">

Figure 1$^{[3]}$: A double pendulum with the lengths, masses, and angles labelled

The kinematic equations of the positions of the double pendulum are$^{[2]}$,

$$x_1=L_1sinθ_1 \: \: \: \: \: \: \: \: \: \: \: \:(1)$$
$$y_1=-L_1cosθ_1 \: \: \: \: \: \: \: \: \: \: \: \:(2)$$
$$x_2=x_1+L_2sinθ_2 \: \: \: \: \: \: \: \: \: \: \: \:(3)$$
$$y_2=y_1-L_2cosθ_2 \: \: \: \: \: \: \: \: \: \: \: \:(4)$$

Where x is the horizontal position of a pendulum, y is the vertical position of a pendulum, L is the length of the rod and θ is the angle of a pendulum, with the subscripts 1 and 2 being the upper and lower pendulum respectively. With the above equations, by taking their derivatives, it is possible to find the velocity of the pendulum with respect to the time of the position$^{[2]}$. This results in the following equations$^{[2]}$,

$$x_1'=θ_1'L_1cosθ_1 \: \: \: \: \: \: \: \: \: \: \: \:(5)$$
$$y_1'=θ_1'L_1sinθ_1 \: \: \: \: \: \: \: \: \: \: \: \:(6)$$
$$x_2'=x_1'+θ_2'L_2cosθ_2 \: \: \: \: \: \: \: \: \: \: \: \:(7)$$
$$y_2'=y_1'+θ_2'L_2sinθ_2 \: \: \: \: \: \: \: \: \: \: \: \:(8)$$

With x’ and y’ being the velocity with respect to time for the respective pendulum and in their respective positions. Again, by taking the derivative of these new equations it is possible to find the acceleration of the pendulum with respect to the time of velocity$^{[2]}$. This results in the following equations$^{[2]}$,

$$x_1'' = −θ_1'^2 L_1 sinθ_1 + θ_1'' L_1 cosθ_1 \: \: \: \: \: \: \: \: \: \: \: \:(9)$$

$$y_1'' = θ_1'^2 L_1 cosθ_1 + θ_1'' L_1 sinθ_1 \: \: \: \: \: \: \: \: \: \: \: \:(10)$$

$$x_2'' = x_1'' − θ_2'^2 L_2 sinθ_2 + θ_2'' L_2 cosθ_2 \: \: \: \: \: \: \: \: \: \: \: \:(11)$$

$$y_2'' = y_1'' + θ_2'2 L_2 cosθ_2 + θ_2'' L_2 sinθ_2 \: \: \: \: \: \: \: \: \: \: \: \:(12)$$

With x’’ and y’’ being the acceleration with respect to time for the respective pendulum and in their respective positions.


Now, using second Newton’s law, which is F = ma, where F is the net force, m is the mass of an object and a is the acceleration, and applying it to the double pendulum. It results in the following equations$^{[2]}$,

$$m_1 x_1'' = −T_1 sinθ_1 + T_2 sinθ_2 \: \: \: \: \: \: \: \: \: \: \: \:(13)$$

$$m_1 y_1'' = T_1 cosθ_1 − T_2 cosθ_2 − m_1 g \: \: \: \: \: \: \: \: \: \: \: \:(14)$$

$$m_2 x_2'' = −T_2 sin θ_2 \: \: \: \: \: \: \: \: \: \: \: \:(15)$$

$$m_2 y_2'' = T_2 cos θ_2 − m_2 g \: \: \: \: \: \: \: \: \: \: \: \:(16)$$

Using equation 10-17, it is now possible to find an expression for $θ_1''$ and $θ_2''$ in terms of $θ_1$, $θ_1'$, $θ_2$, $θ_2'$ through some algebraic manipulations, resulting in the following$^{[2]}$,

$$θ_1''=\frac{−g (2 m_1 + m_2) sin θ_1 − m_2 g sin(θ_1 − 2 θ_2) − 2 sin(θ_1 − θ_2) m_2 (θ_2'^2 L2 + θ_1'^2 L_1 cos(θ_1 − θ_2)} {L_1(2m_1 + m_2 − m_2 cos(2θ_1 − 2θ_2))} \: \: \: \: \: \: \: \: \: \: \: \:(17)$$

$$θ_2'' =\frac{2 sin(θ_1 − θ_2) (θ_1'^2 L_1 (m_1 + m_2) + g(m_1 + m_2) cos θ_1 + θ_2'^2 L_2 m_2 cos(θ_1 − θ_2))} {L_2 (2 m_1 + m_2 − m_2 cos(2 θ_1 − 2 θ_2))} \: \: \: \: \: \: \: \: \: \: \: \:(18)$$

Finally, converting these 2nd order equations and turning them into four 1st order equations, which results in$^{[2]}$,

$$θ_1' = ω_1 \: \: \: \: \: \: \: \: \: \: \: \:(18)$$

$$θ_2' = ω_2 \: \: \: \: \: \: \: \: \: \: \: \:(19)$$

$$ω_1' = \frac{−g (2 m_1 + m_2) sin θ_1 − m_2 g sin(θ_1 − 2 θ_2) − 2 sin(θ_1 − θ_2) m_2 (ω_2^2 L_2 + ω_1^2 L_1 cos(θ_1 − θ_2))} {L_1 (2 m_1 + m_2 − m_2 cos(2 θ_1 − 2 θ_2))} \: \: \: \: \: \: \: \: \: \: \: \:(20)$$

$$ω_2' = \frac{2 sin(θ_1−θ_2) (ω_1^2 L_1 (m_1 + m_2) + g(m_1 + m_2) cos θ_1 + ω_2^2 L_2 m_2 cos(θ_1 − θ_2))} {L_2 (2 m_1 + m_2 − m_2 cos(2 θ_1 − 2 θ_2))} \: \: \: \: \: \: \: \: \: \: \: \:(21)$$

A Lyapunov exponent (λ) is used in the study of dynamical systems to quantitatively measure how trajectories diverge. This allows one to see how stable or predictable a system is. When the exponent is positive, it indicates that the system is chaotic. When it equals 0, it is neutral, and when it is negative, it is stable$^{[4]}$. The larger the exponent, the more chaotic it is, and higher the entropy. The formula for calculating the Lyapunov exponent is given by the equation$^{[4]}$:

$$ \lambda = \lim_{t \to \infty} \frac{1}{t} \ln \frac{\delta(t)}{\delta(0)} \: \: \: \: \: \: \: \: \: \: \: \:(22)$$

Where the initial separation between the trajectories is given by $\delta(0)$.

#II. Model and Numerical Method

The first thing done was to calculate the derivatives of the angular velocity (returns angular accelerations) using the current angles and velocities, which are found using the know initial conditions. The initial conditions are $m_1=m_2=1 kg$ , $L_1=L_2=1 m$ and $g=9.81 m/s^2$, with these being arbritrary values chosen specificaly for this simulation.

Then using the returned angular accelerations to calculate the Lyapunov exponent. It simulates the double pendulum for two different states which vary by a small perturbation (delta=1e-10). The system uses Euler's method to simulate it over time. Euler’s method approximates the slope of the tangent of the function at each step, using the previous point to obtain the next one$^{[1]}$. It then finds the distance between the two states after t=10s to calculate the exponent.

Finally, finding the angles which yields the maximum Lyapunov exponent by simulating a function over a range of initial angles. Where a heatmap is then used to visualize the chaoticness of the double pendulum at different starting angles.

# **Figure 2.** Simluation of motion of double pendulum used for this model
**ignore static image underneath the animation*

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib import rc
import math

# jshtml for displaying animation in colab
rc('animation', html='jshtml')

# Constants
g = 9.81  # gravity
L1 = 1.0  # Rod lengths (m)
L2 = 1.0
m1 = 1.0  # Pendulum masses (kg)
m2 = 1.0

# Initial
theta1 = math.pi / 2  # Initial angles (radians)
theta2 = math.pi / 2
omega1 = 0  # Initial angular velocities (rad/s)
omega2 = 0

# Time
dt = 0.01  # Time step (s) (Lower dt than the other codes due to animation size limit)
t = 10  # Simulation time (s) (increased simulation time for this one to better visualize a double pendulum)

# arrays for storing positions
x1_arr, y1_arr = [], []
x2_arr, y2_arr = [], []

# angular velocities derivatives
def derivatives(theta1, omega1, theta2, omega2):
    delta_theta = theta1 - theta2

    domega1_dt = (-g * (2 * m1 + m2) * math.sin(theta1) - m2 * g * math.sin(theta1 - 2 * theta2) - 2 * math.sin(delta_theta) * m2 * (omega2**2 * L2 + omega1**2 * L1 * math.cos(delta_theta))) / (L1 * (2 * m1 + m2 - m2 * math.cos(2 * delta_theta)))

    domega2_dt = (2 * math.sin(delta_theta) * (omega1**2 * L1 * (m1 + m2)+ g * (m1 + m2) * math.cos(theta1)+ omega2**2 * L2 * m2 * math.cos(delta_theta))) / (L2 * (2 * m1 + m2 - m2 * math.cos(2 * delta_theta)))

    return domega1_dt, domega2_dt

state = np.array([theta1, omega1, theta2, omega2])

# Calculate positions over time
for i in np.arange(0, t, dt):
    theta1, omega1, theta2, omega2 = state
    omega1_dt, omega2_dt = derivatives(theta1, omega1, theta2, omega2)

    theta1 += omega1 * dt
    omega1 += omega1_dt * dt
    theta2 += omega2 * dt
    omega2 += omega2_dt * dt

    x1 = L1 * math.sin(theta1)
    y1 = -L1 * math.cos(theta1)
    x2 = x1 + L2 * math.sin(theta2)
    y2 = y1 - L2 * math.cos(theta2)
    x1_arr.append(x1)
    y1_arr.append(y1)
    x2_arr.append(x2)
    y2_arr.append(y2)

    state = np.array([theta1, omega1, theta2, omega2])

# animation
fig, ax = plt.subplots()
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
line, = ax.plot([], [], 'o-', lw=2)

def init():
    line.set_data([], [])
    return line,

def update(frame):
    line.set_data([0, x1_arr[frame], x2_arr[frame]], [0, y1_arr[frame], y2_arr[frame]])
    return line,

animation = FuncAnimation(fig, update, frames=range(len(x1_arr)), init_func=init, blit=True)

#animation #Disabled due to GitHub importing errors, view on Google Colab.
```

#**Figure 3.** The change of angles of both masses over time starting at an arbitrary angle.

#**Figure 4.** Heat map for the values of Lyapunov's exponent obtained for all combinations of initial angles.

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Constants
g = 9.81  # gravity
L1 = 1.0  # Rod lengths (m)
L2 = 1.0
m1 = 1.0  # Pendulum masses (kg)
m2 = 1.0

# derivatives
def derivatives(theta1, omega1, theta2, omega2):
    delta_theta = theta1 - theta2

    domega1_dt = (-g * (2 * m1 + m2) * math.sin(theta1) - m2 * g * math.sin(theta1 - 2 * theta2) - 2 * math.sin(delta_theta) * m2 * (omega2**2 * L2 + omega1**2 * L1 * math.cos(delta_theta))) / (L1 * (2 * m1 + m2 - m2 * math.cos(2 * delta_theta)))

    domega2_dt = (2 * math.sin(delta_theta) * (omega1**2 * L1 * (m1 + m2)+ g * (m1 + m2) * math.cos(theta1)+ omega2**2 * L2 * m2 * math.cos(delta_theta))) / (L2 * (2 * m1 + m2 - m2 * math.cos(2 * delta_theta)))

    return domega1_dt, domega2_dt

state = np.array([theta1, omega1, theta2, omega2])

# calclating the Lyapunov exponent
def calculate_lyapunov_exponent(theta1_i, theta2_i, omega1_i, omega2_i, dt, t):
    theta1 = theta1_i
    theta2 = theta2_i
    omega1 = omega1_i
    omega2 = omega2_i

    # Small perturbation
    delta = 1e-10
    theta1_p = theta1 + delta

    state = np.array([theta1, omega1, theta2, omega2])
    state_p = np.array([theta1_p, omega1, theta2, omega2])

    max_dist = 0

    for current_time in np.arange(0, t, dt):
        theta1, omega1, theta2, omega2 = state
        theta1_p, omega1_p, theta2_p, omega2_p = state_p

        omega1_dt, omega2_dt = derivatives(theta1, omega1, theta2, omega2)
        omega1_dt_p, omega2_dt_p = derivatives(theta1_p, omega1_p, theta2_p, omega2_p)

        theta1 += omega1 * dt
        omega1 += omega1_dt * dt
        theta2 += omega2 * dt
        omega2 += omega2_dt * dt

        theta1_p += omega1_p * dt
        omega1_p += omega1_dt_p * dt
        theta2_p += omega2_p * dt
        omega2_p += omega2_dt_p * dt

        state = np.array([theta1, omega1, theta2, omega2])
        state_p = np.array([theta1_p, omega1_p, theta2_p, omega2_p])

        # calculating distance b/w two states
        dist = math.sqrt((theta1 - theta1_p)**2 + (theta2 - theta2_p)**2)
        if dist > max_dist:
            max_dist = dist

    # Calculateing Lyapunov exponent
    lyapunov_exponent = np.log(max_dist / delta) / t
    return lyapunov_exponent

# Finding angles with greatest Lyapunov exponent
def find_max_lyapunov_angle(theta_range, omega1_i, omega2_i, dt, t):
    max_lyapunov = 0
    best_theta1 = None
    best_theta2 = None

    lyapunov_values = np.zeros((len(theta_range), len(theta_range)))

    for i, theta1 in enumerate(theta_range):
        for j, theta2 in enumerate(theta_range):
            lyapunov_exp = calculate_lyapunov_exponent(theta1, theta2, omega1_i, omega2_i, dt, t)
            lyapunov_values[i, j] = lyapunov_exp
            if lyapunov_exp > max_lyapunov:
                max_lyapunov = lyapunov_exp
                best_theta1 = theta1
                best_theta2 = theta2

    # Visualization
    plt.figure(figsize=(10, 8))
    plt.imshow(lyapunov_values, extent=(theta_range[0], theta_range[-1], theta_range[0], theta_range[-1]), origin='lower', aspect='auto', cmap='hot')
    plt.colorbar(label='Lyapunov Exponent')
    plt.xlabel('Theta1 (rad)')
    plt.ylabel('Theta2 (rad)')
    plt.title('Lyapunov Exponents for Different Initial Angles')
    plt.plot(best_theta1, best_theta2, 'bo', label=f'Max Lyapunov Exponent = {max_lyapunov:.4f}')
    plt.legend()
    plt.show()

    return best_theta1, best_theta2, max_lyapunov

theta_range = np.linspace(0, 2 * np.pi, 100)
omega1_i = 0
omega2_i = 0
dt = 0.001
t = 5

best_theta1, best_theta2, max_lyapunov = find_max_lyapunov_angle(theta_range, omega1_i, omega2_i, dt, t)
print(f'Best initial angles: Theta1 = {best_theta1}, Theta2 = {best_theta2}')
print(f'Max Lyapunov Exponent: {max_lyapunov}')
```

#**Figure 5.** Validating that the double pendulum model through checking the conservation of energy

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Constants
g = 9.81  # gravity
L1 = 1.0  # Rod lengths (m)
L2 = 1.0
m1 = 1.0  # Pendulum masses (kg)
m2 = 1.0

# Initial
theta1 = math.pi / 2  # Initial angles (radians)
theta2 = math.pi / 2
omega1 = 0  # Initial angular velocities (rad/s)
omega2 = 0

# Time
dt = 0.001  # Time step (s)
t = 5  # Simulation time (s)

# arrays for storing positions & energies
x1_arr, y1_arr = [], []
x2_arr, y2_arr = [], []
KE_arr, PE_arr, TE_arr = [], [], []

# angular velocities derivatives
def derivatives(theta1, omega1, theta2, omega2):
    delta_theta = theta1 - theta2

    domega1_dt = (-g * (2 * m1 + m2) * math.sin(theta1) - m2 * g * math.sin(theta1 - 2 * theta2) - 2 * math.sin(delta_theta) * m2 * (omega2**2 * L2 + omega1**2 * L1 * math.cos(delta_theta))) / (L1 * (2 * m1 + m2 - m2 * math.cos(2 * delta_theta)))

    domega2_dt = (2 * math.sin(delta_theta) * (omega1**2 * L1 * (m1 + m2)+ g * (m1 + m2) * math.cos(theta1)+ omega2**2 * L2 * m2 * math.cos(delta_theta))) / (L2 * (2 * m1 + m2 - m2 * math.cos(2 * delta_theta)))

    return domega1_dt, domega2_dt

# Calculate energy over time
def calculate_energies(th1, w1, th2, w2):
    KE1 = 0.5 * m1 * (L1 * w1)**2
    KE2 = 0.5 * m2 * ((L1 * w1)**2 + (L2 * w2)**2 + 2 * L1 * L2 * w1 * w2 * np.cos(th1 - th2))
    KE = KE1 + KE2

    PE1 = -m1 * g * L1 * np.cos(th1)
    PE2 = -m2 * g * (L1 * np.cos(th1) + L2 * np.cos(th2))
    PE = PE1 + PE2
    TE = KE + PE
    return KE, PE, TE

state = np.array([theta1, omega1, theta2, omega2])

# positions and energies over time
for i in np.arange(0, t, dt):
    theta1, omega1, theta2, omega2 = state
    omega1_dt, omega2_dt = derivatives(theta1, omega1, theta2, omega2)

    theta1 += omega1 * dt
    omega1 += omega1_dt * dt
    theta2 += omega2 * dt
    omega2 += omega2_dt * dt

    x1 = L1 * np.sin(theta1)
    y1 = -L1 * np.cos(theta1)
    x2 = x1 + L2 * np.sin(theta2)
    y2 = y1 - L2 * np.cos(theta2)
    x1_arr.append(x1)
    y1_arr.append(y1)
    x2_arr.append(x2)
    y2_arr.append(y2)

    KE, PE, TE = calculate_energies(theta1, omega1, theta2, omega2)
    KE_arr.append(KE)
    PE_arr.append(PE)
    TE_arr.append(TE)

    state = np.array([theta1, omega1, theta2, omega2])

# Plot energies
plt.figure(figsize=(10, 6))
plt.plot(np.arange(0, t, dt), KE_arr, label='Kinetic Energy')
plt.plot(np.arange(0, t, dt), PE_arr, label='Potential Energy')
plt.plot(np.arange(0, t, dt), TE_arr, label='Total Energy', linestyle='--')
plt.xlabel('Time (s)')
plt.ylabel('Energy (J)')
plt.legend()
plt.title('Energy of the Double Pendulum Over Time (Euler Method)')
plt.show()
```
